# **1. Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

# **2.Data Reading**

```python
data = pd.read_csv('/content/drive/MyDrive/ex1data1.txt',header=None)
data.columns = ['popultion', 'profit']
data.head()
```

```python
x = data.iloc[:, 0].values.reshape(-1,1)
y = data.iloc[:, 1].values.reshape(-1,1)
```

# **3.Data Visulization**

```python
plt.scatter(x,y, color = 'red')
plt.xlabel('Population of city in 10,000s')
plt.ylabel('profit in $10,000s')
plt.show()
```

# **4. Data Splitting & Model training**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33 , random_state=44)
```

```python
model = LinearRegression(fit_intercept= True, normalize= True, copy_X= True)
model.fit(x_train, y_train)
print('model intercept is ', model.intercept_)
print('model cofficient is ', model.coef_)
```

```python
plt.scatter(x_test,y_test, color = 'red')
plt.xlabel('Population of city in 10,000s')
plt.ylabel('profit in $10,000s')
plt.show()
```

# **5. Model Prediction**

```python
y_pred = model.predict(x_test)
print('Treu values of y ', y_test[0:5])
print('predicted values of y ', y_pred[0:5])
```

# **6.Model Visualization**

```python
plt.scatter(x,y, color = 'red')
plt.plot(x_test,y_pred, color = 'blue')
plt.xlabel('Population of city in 10,000s')
plt.ylabel('profit in $10,000s')
plt.show()
```

# **7.Model Evaluation**

```python
import sklearn.metrics as mc
MAEValue = mc.mean_absolute_error(y_test, y_pred, multioutput= 'uniform_average')
print(MAEValue)
```
