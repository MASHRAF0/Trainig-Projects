# **3. Dealing With Null-Nomerical Data**








```python
dataset.isnull().any()
```

```python
dataset['R&D Spend'].median() 
```

```python
dataset['R&D Spend'] = dataset['R&D Spend'].fillna(dataset['R&D Spend'].median())
```

```python
dataset['Administration'].median() 
```

```python
dataset['Administration'] = dataset['Administration'].fillna(dataset['Administration'].median())
```

```python
dataset['Marketing Spend'].median() 
```

```python
dataset['Marketing Spend'] = dataset['Marketing Spend'].fillna(dataset['Marketing Spend'].median())
```

```python
dataset.isnull().any()
```

# **4. Dealing With Null-Categorical Data**


```python
dataset.State.unique()
```

```python
dataset.State.describe()
```

```python
dataset.State.mode()[0]
```

```python
dataset['State'] = dataset['State'].fillna(dataset.State.mode()[0])
```

```python
dataset.isnull().any()
```

# **4. Dealing With Dummy-Categorical Data**

```python
dataset = pd.get_dummies(dataset, columns=['State'] , drop_first = True)
dataset.head()
```

# **5. Data Vlidation**

```python
x = dataset.loc[:,['R&D Spend', 'Administration', 'Marketing Spend']].values
y = dataset['Profit'].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# **6. Data Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaller = StandardScaler()

x_train=scaller.fit_transform(x_train)
x_test=scaller.transform(x_test)

y_train = scaller.fit_transform(y_train)
```

# **7. Model Training**

```python
from sklearn.linear_model import LinearRegression
MLR = LinearRegression()
MLR.fit(x_train, y_train)
```

```python
MLR.coef_ , MLR.intercept_
```

# **8. Model Prediction**

```python
y_pred = MLR.predict(x_test)
```

```python
y_pred = scaller.inverse_transform(y_pred)
```

```python
y_test
```

```python
y_pred
```

# **9. Model Evaluation**

```python
import sklearn.metrics as mc
```

```python
MAEValue = mc.mean_absolute_error(y_test, y_pred, multioutput= 'uniform_average')
print(MAEValue)
```

```python
MSEValue = mc.mean_squared_error(y_test, y_pred, multioutput= 'uniform_average')
print(MSEValue)
```

```python
mc.r2_score(y_test, y_pred, multioutput= 'uniform_average')
```
